```python
with open("dashboard_nucleicos.py", "w") as f:
    pass
    import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
```

```python
# Título del dashboard
st.title('Dashboard Interactivo de Ácidos Nucleicos')
```

```python
# Descripción
st.write("""
Este dashboard interactivo permite visualizar y analizar secuencias de ADN o ARN.
Puedes cargar tu propia secuencia o utilizar una secuencia de ejemplo.
""")
```

```python
# Sección de entrada: carga de secuencia
st.header('Cargar o Ingresar Secuencia de Ácidos Nucleicos')
sequence_input = st.text_area("Ingresa una secuencia de ADN o ARN:")
```

```python
# Visualizar la secuencia
st.subheader('Secuencia ingresada')
st.text(sequence_input)
```

```python
# Convertir la secuencia en un objeto Seq
seq = Seq(sequence_input)
```

```python
# Calcular las estadísticas básicas
longitud = len(seq)
gc_content = gc_fraction(seq) * 100
```

```python
st.write(f"Longitud de la secuencia: {longitud} nucleótidos")
st.write(f"Contenido GC: {gc_content:.2f}%")
```

```python
# Mostrar gráfico de frecuencia de bases
st.subheader('Frecuencia de bases')
base_counts = {
    'A': sequence_input.count('A'),
    'T': sequence_input.count('T'),
    'G': sequence_input.count('G'),
    'C': sequence_input.count('C')
}
```

```python
df_base_counts = pd.DataFrame(list(base_counts.items()), columns=["Base", "Frecuencia"])
```

```python
fig, ax = plt.subplots()
ax.bar(df_base_counts['Base'], df_base_counts['Frecuencia'], color=['red', 'blue', 'green', 'orange'])
ax.set_xlabel('Base')
ax.set_ylabel('Frecuencia')
st.pyplot(fig)
```

```python
# Mostrar la transcripción y traducción si es ADN
if sequence_input.upper().startswith("ATG"):
    st.subheader('Transcripción (ARN)')
    transcripcion = seq.transcribe()
    st.text(str(transcripcion))
```

```python
# Mostrar la transcripción y traducción si es ADN
if sequence_input.upper().startswith("ATG"):
    st.subheader('Transcripción (ARN)')
    transcripcion = seq.transcribe()
    st.text(str(transcripcion))

    # This line should be indented to be part of the 'if' block
    st.subheader('Traducción (Proteína)')
    traduccion = seq.translate()
    st.text(str(traduccion))
```
